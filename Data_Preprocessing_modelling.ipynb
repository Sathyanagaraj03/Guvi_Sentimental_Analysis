```python
df.drop(columns=["id","entity"],inplace=True)
```

```python
df['comment']=df['comment'].str.lower()
df.head()
```

```python
import unicodedata

def has_emoji(text_list):
    for text in text_list:
        for char in text:
            if char != ' ' and unicodedata.name(char).startswith('EMOJI'):
                return True
    return False

# Apply the function to the 'sentiment' column
df['emoji'] = df['comment'].apply(has_emoji)

df.head()
```

```python
df = df[df["emoji"] == False]
```

```python
def remove_whitespace(text):
    return  " ".join(text.split())


df['comment']=df['comment'].apply(remove_whitespace)
df.head()
```

```python
import nltk
nltk.download('punkt')
```

```python
from nltk import word_tokenize
df['comment']=df['comment'].apply(lambda X: word_tokenize(X))
df.head()
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords

en_stopwords = stopwords.words('english')


def remove_stopwords(text):
    result = []
    for token in text:
        if token not in en_stopwords:
            result.append(token)

    return result
```

```python
df['comment']=df['comment'].apply(remove_stopwords)
df.head()
```

```python
from nltk.tokenize import RegexpTokenizer

def remove_punct(text):

    tokenizer = RegexpTokenizer(r"\w+")
    lst=tokenizer.tokenize(' '.join(text))
    return lst


df['comment']=df['comment'].apply(remove_punct)
df.head()
```

```python
def alphabetical_only(comment_list):
    return [word for word in comment_list if word.isalpha()]

# Apply the function to the 'sentiment' column
df['comment'] = df['comment'].apply(alphabetical_only)
df.head()
```

```python
def remove_im(comment_list):
    return [word for word in comment_list if word != 'im']

# Apply the function to the 'sentiment' column
df['comment'] = df['comment'].apply(remove_im)
df.head()
```

```python
# Convert lists to tuples in the 'sentiment' column
df['comment'] = df['comment'].apply(tuple)

# Drop duplicates
df = df.drop_duplicates()

# Convert tuples back to lists (if necessary)
df['comment'] = df['comment'].apply(list)

# Update the len_sent_1 column to reflect the new length of sentiment lists
# df['len_sent_1'] = df['sentiment'].apply(len)

# Check the shape after dropping duplicates
print("Shape after dropping duplicates:", df.shape)

# Print the updated DataFrame
# print(df)
```

```python
def remove_single_letters(text):
    return [word for word in text if len(word) > 1]

# Apply the function to the 'sentiment' column
df['comment'] = df['comment'].apply(remove_single_letters)
     

df["comment_length"] = df['comment'].apply(lambda X: len(X))
df.head()
```

```python
from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize,pos_tag
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

def lemmatization(text):

    result=[]
    wordnet = WordNetLemmatizer()
    for token,tag in pos_tag(text):
        pos=tag[0].lower()

        if pos not in ['a', 'r', 'n', 'v']:
            pos='n'

        result.append(wordnet.lemmatize(token,pos))

    return result


df['comment']=df['comment'].apply(lemmatization)
df.head()
```

```python
df['Comments'] = df['comment'].apply(lambda x: ' '.join(x))
df.head()
```

```python
df.to_csv("dataset.csv")
```
